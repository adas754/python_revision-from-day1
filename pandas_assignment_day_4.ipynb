###### 18. Write a Pandas program to get sample 75% of the diamonds DataFrame's rows without replacement and store the remaining 25% of the rows in another DataFrame.

```python
sample_df=df.sample(frac=0.75,replace=False,random_state=42)
remaining_df=df.loc[~df.index.isin(sample_df.index)]
print(f'Sampled diamonds: {len(sample_df)} rows')
print(f'Remaining diamonds : {len(remaining_df)} rows')
```
