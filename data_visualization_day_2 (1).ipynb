##### 1. Write a Python program to display a bar chart of the fare of class.

```python
### code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
class_fairs=df.groupby('pclass')['fare'].mean()
class_fairs.plot(kind='bar')
plt.title('Average Fare by class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here
class_fairs=df.groupby('pclass')['fare'].mean()
class_fairs.plot(kind='barh')
plt.title('Average fare by class')
plt.xlabel('Fare')
plt.ylabel('class')
plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
class_fares=df.groupby('pclass')['fare'].mean()
colors=['red','green','blue']
class_fares.plot(kind='bar',color=colors)
plt.title('Average Fare by class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
### code here
class_fare=df.groupby('pclass')['fare'].sum()
class_fare.plot(kind='pie',autopct='%1.1f%%')
plt.title('total fare by class')

plt.show()
```

######5. Write a Python program to display a count plot of all categorical columns.

```python
### code here
for col in df.columns:
  if df[col].dtype=='object':
    sns.countplot(data=df,x=col)
    plt.title(f'count plot of {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

```python
  #### code here
sns.catplot(df,x='pclass',col='embarked',kind='count')
plt.show()
```

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
### code here
for col in df.columns:
  if df[col].dtype=='int64':
    sns.countplot(data=df,x=col)
    plt.title(f'count plot of {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
```
