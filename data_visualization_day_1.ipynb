######1. load the iris.csv dataset.

```python
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
df.head()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
import seaborn as sns
import matplotlib.pyplot as plt
species_count=df['Species'].value_counts()
sns.barplot(x=species_count.index,y=species_count.values)
plt.xlabel('Species')
plt.ylabel('count')
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
plt.pie(species_count.values, labels=species_count.index,autopct='%1.1f%%')
plt.title('percentage of species')
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=df,kind='reg')
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',hue="Species",data=df)
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
sns.boxplot(x='SepalLengthCm',data=df)
plt.show()
```

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=df)
plt.show()
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(x='Species',y='PetalLengthCm',data=df)
plt.show()
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(y='Species',x='SepalLengthCm',data=df)
plt.show()
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(data=df,hue='Species')
plt.show()
```

##### 9. Create a Heat Map

```python
### code here
corr=df.corr()
sns.heatmap(corr,cmap='coolwarm',annot=True)
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
sns.displot(data=df,kind='kde')
plt.show()
```
